### Apply

```python
# split into lists
names_list=[]
percent_list=[]
for i in range(len(lines)):
    # get rid of \n at the end
    line = lines[i][:-1].split("\t")
    # append the first element to names_list
    names_list.append(line[0])
    # append the rest elements to percent_list
    percent_list.append(line[1:])
```

```python
len(names_list)
```

```python
len(percent_list)
```

```python
# make the lists into Tableau friendly forms
names_tab=[]
percent_tab=[]
```

```python
for i in range(len(names_list)):
    names_tab += [names_list[i]]*len(percent_list[i])
    percent_tab += percent_list[i]
```

```python
# create a dataframe
df = pd.DataFrame(columns=["name","percent"])
df.name = names_tab
df.percent = percent_tab
```

```python
df
```

```python
# change data type
df.percent = df.percent.apply(lambda x: float(x))
```

```python
# export
df.to_csv("Boder_Ngram_Pseudo_Index.csv",index=False)
```
