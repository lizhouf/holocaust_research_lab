### From XML

```python
tree_8_1 = ET.parse('8.1.xml')
root_8_1 = tree_8_1.getroot()
tree_8_2 = ET.parse('8.2.xml')
root_8_2 = tree_8_2.getroot()
tree_8_3 = ET.parse('8.3.xml')
root_8_3 = tree_8_3.getroot()
tree_8_4 = ET.parse('8.4.xml')
root_8_4 = tree_8_4.getroot()
tree_8_5 = ET.parse('8.5.xml')
root_8_5 = tree_8_5.getroot()
```

```python
def print_xml(root_file):
    start_minisecond = 0
    is_reduce = False
    for span in root_file.iter('span'):
        # start minisecond
        if is_reduce == False:
            start_minisecond = int(span.attrib['m'])
            is_reduce = True
        # calculate
        miniseconds = int(span.attrib['m'])-start_minisecond # total minisecond
        hour = miniseconds//3600000 # the hour
        minute = (miniseconds-hour*3600000)//60000 # the minute
        second = (miniseconds-hour*3600000-minute*60000)//1000 # the second
        minisecond = miniseconds-hour*3600000-minute*60000-second*1000 # the minisecond
        print(span.attrib['m'],str(hour)+":"+str(minute)+":"+str(second)+":"+str(minisecond),span.text)
```

```python
# the first xml file: 29 segments
print_xml(root_8_1)
```
